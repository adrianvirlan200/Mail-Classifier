```python
import  numpy as np
import os
import string
import sys
from matplotlib import pyplot as plt 

def LQ(A): #triangularizarea la dreapta cu reflectori hermitici 
    m,n  = np.shape(A)
    s = min(m,n)
    
    beta = np.zeros(s)

    for k in range(0,s,1):
        beta[k] = 0
        if k<n:
            sigma = np.linalg.norm(A[k,k:n],2)
            if sigma!=0:
                if A[k,k] !=0:
                    sigma = sigma * A[k,k]/np.absolute(A[k,k])
                for j in range(k,n,1):
                    A[k,j] = A[k,j]/sigma
                beta[k] = 1 + A[k,k]
                for i in range(k+1,m,1):
                    alfa = 0
                    for j in range(k,n,1):
                        alfa += A[i,j] * A[k,j]
                    alfa = -1*alfa/beta[k]
                    for j in range(k,n,1):
                        A[i,j] += alfa*A[k,j]
                A[k,k] = -sigma
    return A, beta
    
def SQL(A, b, beta): #rezolvarea sistemelor subdeterminate
    m,n  = np.shape(A)
    x = np.zeros(n)
    
    for k in range(0,m,1):
        sum = 0
        for j in range(0, k-1, 1):
            sum += A[k,j]*x[j]
        x[k] = (b[k] - sum)/A[k,k]
    #calculul solutiei normale
    x[m+1:n] = 0
    for k in range(m-1,-1,-1):
        t = A[k,k]
        A[k,k] = beta[k]
        sum = 0
        for j in range(k,n,1):
            sum+=A[k,j]*x[j]
        alfa = -sum/beta[k]
        for j in range(k,n,1):
            x[j]+=alfa*A[k,j]
        A[k,k] = t
    return x
        
```

```python
#citeste cuvintele din dictionar in vectorul "dictionar"
dictionar = np.loadtxt("dictionar2.txt", dtype=str)
#d = dimensiune vector dictionar
d = dictionar.size

#salveaza numele fiecarui fisier pentru a fi citite mail-urile de antrenare
mail_bun = ["1.txt","2.txt","3.txt","4.txt","5.txt"]
mail_spam = ["6.txt","7.txt","8.txt","9.txt","10.txt"]
#in total 10 mail-uri de antrenare
nr_mailuri_antrenare = 10 

#initializeaza matrice features(dim 10 X d+1)
features = np.zeros((nr_mailuri_antrenare, d + 1))   #ultima coloana = 1

#initializeaza vectorul labels de dim 10 X 1(10 mailuri, fiecare mail este bun(1) sau spam(-1))
labels = np.zeros(nr_mailuri_antrenare)


nr_mail = -1
#5 mail-uri bune de antrenare
for i in mail_bun:
    nr_mail += 1
    
    features[nr_mail, d] = 1 #ultima linie este 1
    labels[nr_mail] = 1      #mail bun (1)               
    
    with open(i,'r') as file:   #deschide fisierul "nr_mail.txt"
        
        # citeste fiecare linie    
        for line in file:
        
            # citeste fiecare cuvant din mail        
            for word in line.split():
                
                #cauta daca cuvantul se gaseste in dictionar
                 for j in range(0, d, 1):               
                        if(word.lower() == dictionar[j]): #daca cuvantul este gasit
                            features[nr_mail, j] += 1 #incrementeaza cu 1 nr de aparitii
                            break #iesi din for
                            

#5 mail-uri spam de antrenare
for i in mail_spam:
    nr_mail += 1
   
    features[nr_mail, d] = 1 #ultima linie este 1
    labels[nr_mail] = -1      #mail spam (-1)       
   
    with open(i,'r') as file:   #deschide fisierul "nr_mail.txt"
        
        # citeste fiecare linie    
        for line in file:

            # citeste fiecare cuvant din mail        
            for word in line.split():
                 
                 #cauta daca cuvantul se gaseste in dictionar
                 for j in range(0, d, 1):                 
                        if(word.lower() == dictionar[j]):#daca cuvantul este gasit
                            features[nr_mail, j] += 1 #incrementeaza nr de aparitii
                            break #iesi din for


#features
#labels

features = np.unique(features, axis=0)
```

```python
#testam:

mail_test = ["mail_bun_test.txt","mail_bun_test2.txt","mail_spam_test.txt","mail_spam_test2.txt"]
test_features = np.zeros([4,d])

nr_mail = -1;
for i in mail_test:
    nr_mail += 1 
    with open(i,'r') as file:   #deschide fisierul cu mailul de testare

            # citeste fiecare linie    
            for line in file:

                # citeste fiecare cuvant din mail        
                for word in line.split():

                    #cauta daca cuvantul se gaseste in dictionar
                     for j in range(0, d, 1):                       
                            if(word.lower() == dictionar[j]):
                                test_features[nr_mail,j] += 1 #nr de aparitii
                                break

    
```

```python
# sistemul este supradeterminat => cautam solutia CMMP
A = np.copy(features)

x1 = np.linalg.lstsq(features ,labels, rcond=None)[0]

#clasificatorul obtinut este de forma: f(v) = c' * v + d
#la noi x[:d] = c si c[d] = d 
                            
                            
                            
#x2 = np.transpose(features) @ np.linalg.inv(features @ np.transpose(features)) @ labels

L, beta = LQ(features)
x2 = SQL(features, labels, beta)
    
# grafic cu diferentele dintre apelarea lstsq si apelarea LQ + SQL
plt.plot(x1 - x2)
plt.legend(["Solutie lstsq(x1) - Solutie apelare LQ si SQL(x2)"])
plt.show()

#x1 = np.copy(x2)


# clasificare mailuri antrenament
for i in range(0,10,1):
    label_mail = A[i,:d] @ x1[:d] + x1[d] #aplica clasificatorul pe mailul i
    plt.plot(label_mail,'*', label='mail antrenament'+ str(i+1), markersize=10)
    print('mail_antrenament' + str(i+1) + str(": "), label_mail)

plt.xlabel("label_obtinut")
plt.legend()

plt.plot(x1,x1, label='separator de clase')

#scrie legenda in afara graficului
ax = plt.subplot(111)
# Shrink current axis by 20%
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
    
    
#clasificare mailuri noi
print('----test_mails----')
for i in range(0, 4, 1):
    label_mail = test_features[i,:] @ x1[:d] + x1[d]#aplcia  clasificatorul
    if i <= 1:
        string = 'bun'
    else:
        string = 'spam'
    plt.plot(label_mail, 'x' ,label='mail_test_'+ string)
    print('mail de test' + str(i+1),label_mail)
    
plt.plot(x1,x1, label='separator de clase')

plt.legend()
plt.xlabel("label obtinut")
#scrie legenda in afara graficului
ax = plt.subplot(111)
# Shrink current axis by 20%
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

 
```
